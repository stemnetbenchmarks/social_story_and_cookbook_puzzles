# How Spells and Elements can work in a Five-Element Resonance System:

# Spells

In general, overall, in broadstrokes, in general principles, in Dungeon-Of-Notebooks, spells and elemental allignments for Dungeon-of-Notebooks work ~similiarly to the traditional eastern fire-element system where each element helps or harms other elements.

For example: In Dungeon-of-Notebooks, you cannot simply barge into any room and let fly any spell and expect a generic predictable effect such as: x points of "general" damage, which somehow effects everything the same exact way.

Effects will depend on elements; a fire-ball may cause damage...or it might help, or it might have no effect at all. The same for any elemental-category of spell.

# Modifying the Traditional five-element system

We are using a five-elment system because it is a wonderful model of differing cross-effects. Everything effects something else positively and something else negatively (and like-wise is affected positively and negatively by other things).

We are using a 'modified' version because a lot of the 'help' or 'harm' relationships in the original system do not make enough intutitive or rational sense in a concrete way for our purposes here. We are not studying abstract relationships in eastern medicine, we want concrete relationships for spells in a story game.

Arguably the best parts of the orignal system are that water harms fire, and wood helps fire. But most of the other relationships are either nonsense or arbitrary stretches of rationalization that could be argued in any direction.

Specifically, these are the dubious relationships from the original system.

## Relationships that do not make enough sense:
- Earth harms water (earth soil and rocks are integral to the water cycle)
- Wood harms earth (Wood in no way harms soil or rocks...or planets.)
- fire harms metal (Metal only exists as such because of fire.)
- metal harms wood (Metal in a forest will be hapily ignored. 'Metal' elements are required for the chemistry and biochemistry of ecosystems.)

 &

- fire helps earth (This is a rationalization stretch that requires extremely selective marginal examples. If a fire-regime is carefully in balance it does not harm the ecosystem, but 'more fire = better' is false. Some geology requires 'fire' at some points (very selectively), but randomly adding fire to all crystalization, chemistry, and geology, will be negative the vast majority of the time. And generally, rocks are insulators not reactive. Chimneys and firepits are made with rocks because that fire has no effect on the rocks.)
- earth helps metal (Adding rocks and dirt to metal won't help the metal and will probably damage it.)
- metal helps water (Most metals are not something you want to add to your drinking water. Without a context this is a nonsensical statement.)
- water helps wood (Water helps living plants, but it is BAD for wood.)

#### 2: Metal has been changed to --> Ice

# New Elements & Relationships

Helps List:
- Water +> Forest (add water to a Forest Ecosystem, you get more trees etc.)
- Forest +> Fire (add wood to a fire, you get more fire)
- Fire +> Void (add fire to a full space, you get more empty space)
- Void +> Ice ('add' lower pressure, you get more ice. (PV = nRT) creating more vaccume (lowering pressure) lowers temperature to make more ice, and vaccume is literally the only good insulator to help maintain ice; this is how refrigerators work, this is really how ice is made)
- Ice +> Water (add ice to water -> you get more water)

Harms List:
- Water -> Fire (add water to fire, you get less fire)
- Fire -> Ice (add fire to ice, you get less ice)
- Ice -> Forest (add ice to plants, you get slower-growth or dead plants (a.k.a "Winter"))
- Forest -> Void (add growing ecosystems into empty space, you get less empty space (This is less rigorous, but hopefull makes sense.))
- Void -> Water (Two approaches to this: 1. Add 'lower-pressure' and water turns to ice (less water, more ice). 2. Dig a trench to a body of water, and that body of water shrinks: Gasses and liquids move from high concentration to low concentration, from high pressure to low pressure.)(Maybe Less rigorously, another example: 3. The vaccume of space, unimpeeded, will strip atomosphere and liquids from a body in space: asteroids are 'bare rock' or ice so cold it does not behave a gas or liquid.)

Not perfect, but hopefully better.

# Absorption

As another game-mechanic to avoid unilateral actions, an object can be 'absorbed' into the 'inventory' of another object. There are various ways this could work (and the details may be arbitrary).

Exmple:
If object-A helps another object-B, but object-A has no energy,
then object-B can 'absorb, carry, maintain' object-A as an inventory-item.

Three RPG examples of this are:
1. How do you add a treasure to inventory?
2. How do you pull a defeated monster/npc into inventory?
3. How do you carry a wounded team-member to help heal them later?

All these can follow naturall from a simple 'absorb into inventory' rule.

#### Note:
It is possible to have a 'unilateral' system where harming something else automatically leads to absorption (having the rule be the being able to harm something enables you to collect that into inventory), but this is both contrary to the overall point of having 'cut-up' inter-related roles (team collaboration) and intuitively makes little sense: adding something to inventory is not an attack that damages that object; destroying and object and preserving it for later use are not consistent with each other.
